```python
import pandas as pd
```

```python
# Importing the dataset using pandas
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
# Displaying the first five rows of the dataset
df.head()
```

```python
# Basic structure of the dataset
df.info()
```

```python
# Statistical summary of numerical columns
df.describe()
```

```python
# Visualizing missing values using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Average Purchase by Gender
df.groupby('Gender')['Purchase'].mean().plot(kind='bar', color=['blue', 'pink'])
plt.title('Average Purchase by Gender')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
# Average Purchase by Age Group
df.groupby('Age')['Purchase'].mean().plot(kind='bar', color='orange')
plt.title('Average Purchase by Age Group')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
# Average Purchase by Product Category 1
df.groupby('Product_Category_1')['Purchase'].mean().plot(kind='bar', color='green')
plt.title('Average Purchase by Product Category 1')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
# Displaying the number of unique values in each column
for column in df.columns:
    print(df[column].nunique(), "\t: ", column)
```

```python
# Creating a pie chart to show gender distribution of customers
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]}, 
                    index = ['Male', 'Female'])

data.plot.pie(y = 'Ratio', figsize = (6,6), autopct = "%.1f")
```
